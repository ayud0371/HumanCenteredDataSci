First, we import pandas, matplotlib, and seaborn to use as we analyze our data.
Then, we load our datasets (us_COVID_cases_and_deaths_by_state.csv and us_COVID_vaccinations_by_state.csv) into Python as a PANDAS dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
cases_and_deaths_df = pd.read_csv('us_COVID_cases_and_deaths_by_state.csv')
cases_and_deaths_df.head()
vaccinations_df = pd.read_csv('us_COVID_vaccinations_by_state.csv')
vaccinations_df.head()
```

Now I want to clean and plot the correlation between these two data sets.
Here, I am only taking the columns I need.

```python
cases_and_deaths_df_filtered = cases_and_deaths_df[['date','state', 'tot_death', 'tot_cases']]
cases_and_deaths_df_filtered.head()
vaccinations_df_filtered = vaccinations_df[['date','state', 'people_vaccinated']]
vaccinations_df_filtered.head()
```

For the state column, we have strings as our state descriptions. If we want to use it as an input to our model later on, it needs to be a numeric value. (1 means that the row belongs to the state and 0 means that the row doesn't)

```python
cases_and_deaths_final = cases_and_deaths_df_filtered.join(pd.get_dummies(cases_and_deaths_df_filtered['state']))
cases_and_deaths_final.head()
vaccinations_final = vaccinations_df_filtered.join(pd.get_dummies(vaccinations_df_filtered['state']))
vaccinations_final.head()
# We only want to see the data for Texas
cases_final = cases_and_deaths_final[['date', 'tot_cases', 'TX']]
deaths_final = cases_and_deaths_final[['date', 'tot_death', 'TX']]
vaccinations_final = vaccinations_final[['date', 'people_vaccinated', 'TX']]
```

Now we can create a visualization that shows the correlation between COVID vaccination and cases/deaths in Texas

```python
# histograms!
c_TX = cases_final.loc[(cases_final.TX == 1) , 'date']
d_TX = deaths_final.loc[(deaths_final.TX == 1) , 'date']
v_TX = vaccinations_final.loc[(vaccinations_final.TX == 1) , 'date']

plt.subplot(1, 2, 1)

# density=True would normalize the distribution, but it's a little harder to see the separate categories
kwargs = dict(stacked=True, histtype='step')

plt.xlabel('date')
plt.ylabel('Number Of Individuals')
plt.title = ('COVID-19 CASES, DEATHS, AND VACCINATIONS Distribution')
plt.hist(c_TX, label='cases', color='r', **kwargs)
plt.hist(d_TX, label='deaths', color='g', **kwargs)
plt.hist(v_TX, label='vaccinations', color='b', **kwargs)

# plt.xlim(1,1100)
# plt.ylim(0,100)
plt.rcParams["figure.figsize"] = (100,60)
plt.legend()

plt.show()
```
